```python
import numpy as np
filename = 'Sample_Input_1_2c.txt'

def initialize_dictionary(k):
    return [i for i in range(1, k + 1)]

dictionary = initialize_dictionary(n)

def initialize_pos(pi):
    r = np.random.random()
    cumulative_distribution = np.cumsum(pi)
    for q, probability in enumerate(cumulative_distribution):
        if r <= probability:
            return q

def textGen(pi, A, B, tl):
    text = []
    current_state = initialize_pos(pi)
    for _ in range(tl):
        next_state = np.random.choice(len(pi), p=A[current_state])
        number = np.random.choice(dictionary, p=B[next_state].flatten())
        text.append(number)
        current_state = next_state
    return text

def main(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    first_line = lines[0].split()
    m = int(first_line[0])
    n = int(first_line[1])
    tl = int(first_line[2])
    pi = list(map(float, lines[1].split()))

    A = []
    for i in range(m):
        A.append(list(map(float, lines[2 + i].split())))

    B = []
    for i in range(m):
        B.append(list(map(float, lines[2 + m + i].split())))

    np.random.seed(0)
    output = textGen(pi, A, B, tl)
    print(output)

main(filename)






```
